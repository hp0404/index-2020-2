```python
import numpy as np
import pandas as pd
from custom_func import wa
```

```python
PATH = "./../v0.5/outputs/calculated_index"
```

```python
P1 = pd.read_csv(f"{PATH}/P1.csv")
P2 = pd.read_csv(f"{PATH}/P2.csv")
P3 = pd.read_csv(f"{PATH}/P3.csv")
P4 = pd.read_csv(f"{PATH}/P4.csv")
P5 = pd.read_csv(f"{PATH}/P5.csv")
P6 = pd.read_csv(f"{PATH}/P6.csv")
P7 = pd.read_csv(f"{PATH}/P7.csv")
P8 = pd.read_csv(f"{PATH}/P8.csv")
```

```python
df_P = pd.DataFrame(P1['region'])

P_sources = [P1, P2, P3, P4, P5, P6, P7, P8]
for source in P_sources:
    df_P = df_P.merge(source.loc[:, source.columns.str.contains("region$|P")], on="region", how="outer")
```

```python
df_P["Score"] = wa(df_P.fillna(0))
df_P.sort_values("Score", ascending=False).reset_index(drop=True)
```
